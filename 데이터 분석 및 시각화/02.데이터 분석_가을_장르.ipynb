## 데이터 추출

```python
import pandas as pd

# 파일 읽기
month_2017 = pd.read_excel('./data/2017_month.xlsx')
month_2018 = pd.read_excel('./data/2018_month.xlsx')
month_2019 = pd.read_excel('./data/2019_month.xlsx')
month_2020 = pd.read_excel('./data/2020_month.xlsx')
month_2021 = pd.read_excel('./data/2021_month.xlsx')
```

```python
# 2017~2020 가을(9, 10, 11월) 데이터 추출
Fall = pd.DataFrame()
for data in [month_2017, month_2018, month_2019, month_2020]:
    Fall = pd.concat([Fall, data[data['월'].isin([9, 10, 11])]], ignore_index=True)
Fall
```

```python
# 엑셀파일로 저장
columns = ['년도', '월', '순위', '타이틀', '가수', '장르']
pd_data = pd.DataFrame(Fall, columns = columns)

pd_data.to_excel('./data/5years_fall_chart.xlsx', index=False)
```

```python
# 2017~2020 9월 데이터 추출
Sep = Fall[Fall['월']==9]
Sep.reset_index(drop=True, inplace=True)
Sep
```

```python
# 2017~2020 10월 데이터 추출
Oct = Fall[Fall['월']==10]
Oct.reset_index(drop=True, inplace=True)
Oct
```

```python
# 2017~2020 11월 데이터 추출
Nov = Fall[Fall['월']==11]
Nov.reset_index(drop=True, inplace=True)
Nov
```

```python
# 연도별 가을(9~11월) 데이터 추출
Fall_2017 = Fall[Fall['년도']==2017]
Fall_2018 = Fall[Fall['년도']==2018]
Fall_2019 = Fall[Fall['년도']==2019]
Fall_2020 = Fall[Fall['년도']==2020]
```

```python
# 연도별 9월 데이터 추출
Sep_2017 = Sep[Sep['년도']==2017]
Sep_2018 = Sep[Sep['년도']==2018]
Sep_2019 = Sep[Sep['년도']==2019]
Sep_2020 = Sep[Sep['년도']==2020]
```

```python
# 연도별 10월 데이터 추출
Oct_2017 = Oct[Oct['년도']==2017]
Oct_2018 = Oct[Oct['년도']==2018]
Oct_2019 = Oct[Oct['년도']==2019]
Oct_2020 = Oct[Oct['년도']==2020]
```

```python
# 연도별 11월 데이터 추출
Nov_2017 = Nov[Nov['년도']==2017]
Nov_2018 = Nov[Nov['년도']==2018]
Nov_2019 = Nov[Nov['년도']==2019]
Nov_2020 = Nov[Nov['년도']==2020]
```

## 장르 분포 분석

### 연도별 가을

```python
# 2017년 가을 장르 개수
GenreCount_2017 = pd.DataFrame([Sep_2017['장르'].value_counts(), Oct_2017['장르'].value_counts(), 
                                Nov_2017['장르'].value_counts(), Fall_2017['장르'].value_counts()],
                                index=['9월', '10월', '11월', '합계'])
GenreCount_2017 = GenreCount_2017.T
GenreCount_2017.sort_values('합계', ascending=False, inplace=True)
GenreCount_2017
```

```python
# 2017~2020 9~11월 장르 개수
GenreCount_Fall = pd.DataFrame([Sep['장르'].value_counts(), Oct['장르'].value_counts(),
                           Nov['장르'].value_counts(), Fall['장르'].value_counts()],
                           index=['9월', '10월', '11월', '합계'])
GenreCount_Fall = GenreCount_Fall.T
GenreCount_Fall.sort_values('합계', ascending=False, inplace=True)
GenreCount_Fall
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
# 한글 표시
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False

# 색상 지정
colors = ['indianred', 'coral', 'khaki', 'lightgreen', 'lightblue', 'steelblue', 'plum', 'lightgrey', 'darkseagreen', 'mediumaquamarine', 'lightpink']
```

### 2017~2020년 합계

```python
GenreCount_Fall.plot(kind='barh', y=['9월', '10월', '11월'], color=colors, figsize=(15,5), stacked=True)
plt.title("2017~2020년 가을 장르 분포")
plt.xlabel("개수")
plt.ylabel("장르")
```

```python
plt.figure(figsize=(15,15)).set_facecolor('white')
# 2017~2020년 가을
plt.subplot(2,2,1)
plt.pie(GenreCount_Fall['합계'], autopct='%.1f%%', startangle=90, counterclock = False, colors=colors)
plt.title('2017~2020년 가을(9월~11월) 장르 분포', fontsize=15)
plt.legend(labels=GenreCount_Fall.index, title='장르', bbox_to_anchor=(1, 0.5))

# 2017~2020년 9월
plt.subplot(2,2,2)
plt.pie(GenreCount_Sep['합계'], autopct='%.1f%%', startangle=90, counterclock = False, colors=colors)
plt.title('2017~2020년 9월 장르 분포', fontsize=15)
plt.legend(labels=GenreCount_Sep.index, title='장르', bbox_to_anchor=(1, 0.5))

# 2017~2020년 10월
plt.subplot(2,2,3)
plt.pie(GenreCount_Oct['합계'], autopct='%.1f%%', startangle=90, counterclock = False, colors=colors)
plt.title('2017~2020년 10월 장르 분포', fontsize=15)
plt.legend(labels=GenreCount_Oct.index, title='장르', bbox_to_anchor=(1, 0.5))

# 2017~2020년 11월
plt.subplot(2,2,4)
plt.pie(GenreCount_Nov['합계'], autopct='%.1f%%', startangle=90, counterclock = False, colors=colors)
plt.title('2017~2020년 11월 장르 분포', fontsize=15)
plt.legend(labels=GenreCount_Nov.index, title='장르', bbox_to_anchor=(1, 0.5))
```
