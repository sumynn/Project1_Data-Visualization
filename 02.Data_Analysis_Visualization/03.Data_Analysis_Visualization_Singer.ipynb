## 데이터 추출

```python
import pandas as pd

# 파일 읽기
month_2017 = pd.read_excel('./data/2017_month.xlsx')
month_2018 = pd.read_excel('./data/2018_month.xlsx')
month_2019 = pd.read_excel('./data/2019_month.xlsx')
month_2020 = pd.read_excel('./data/2020_month.xlsx')
month_2021 = pd.read_excel('./data/2021_month.xlsx')
```

```python
# 2017~2020 가을(9, 10, 11월) 데이터 추출
Fall = pd.DataFrame()
for data in [month_2017, month_2018, month_2019, month_2020]:
    Fall = pd.concat([Fall, data[data['월'].isin([9, 10, 11])]], ignore_index=True)
Fall
```

```python
# 2017~2020 9월 데이터 추출
Sep = Fall[Fall['월']==9]
Sep.reset_index(drop=True, inplace=True)
Sep
```

```python
# 2017~2020 10월 데이터 추출
Oct = Fall[Fall['월']==10]
Oct.reset_index(drop=True, inplace=True)
Oct
```

```python
# 2017~2020 11월 데이터 추출
Nov = Fall[Fall['월']==11]
Nov.reset_index(drop=True, inplace=True)
Nov
```

```python
# 연도별 가을(9~11월) 데이터 추출
Fall_2017 = Fall[Fall['년도']==2017]
Fall_2018 = Fall[Fall['년도']==2018]
Fall_2019 = Fall[Fall['년도']==2019]
Fall_2020 = Fall[Fall['년도']==2020]
```

```python
# 연도별 9월 데이터 추출
Sep_2017 = Sep[Sep['년도']==2017]
Sep_2018 = Sep[Sep['년도']==2018]
Sep_2019 = Sep[Sep['년도']==2019]
Sep_2020 = Sep[Sep['년도']==2020]
```

```python
# 연도별 10월 데이터 추출
Oct_2017 = Oct[Oct['년도']==2017]
Oct_2018 = Oct[Oct['년도']==2018]
Oct_2019 = Oct[Oct['년도']==2019]
Oct_2020 = Oct[Oct['년도']==2020]
```

```python
# 연도별 11월 데이터 추출
Nov_2017 = Nov[Nov['년도']==2017]
Nov_2018 = Nov[Nov['년도']==2018]
Nov_2019 = Nov[Nov['년도']==2019]
Nov_2020 = Nov[Nov['년도']==2020]
```

### 2017~2020 월별 가수 차트인 횟수

```python
# 2017~2020 가수별 차트인 횟수
SingerCount = pd.DataFrame([month_2017['가수'].value_counts(), month_2018['가수'].value_counts(),
                            month_2019['가수'].value_counts(), month_2020['가수'].value_counts()])
SingerCount = SingerCount.append(SingerCount.sum(), ignore_index=True).T
SingerCount.columns =['2017년', '2018년', '2019년', '2020년', '합계']
SingerCount.sort_values('합계', ascending=False)
```

### 2017~2020년 가을 총 합계

```python
# 2017~2020 9~11월 가수별 차트인 횟수
SingerCount_Fall = pd.DataFrame([Sep['가수'].value_counts(), Oct['가수'].value_counts(),
                                 Nov['가수'].value_counts(), Fall['가수'].value_counts()],
                                 index=['9월', '10월', '11월', '합계'])
SingerCount_Fall = SingerCount_Fall.T
SingerCount_Fall.sort_values('합계', ascending=False)
```

## 시각화

```python
import matplotlib.pyplot as plt
# 한글 표시
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False

# 색상 지정
colors = ['indianred', 'coral', 'khaki', 'lightgreen', 'lightblue', 'steelblue', 'plum', 'lightgrey', 'darkseagreen', 'mediumaquamarine', 'lightpink']
```

### 2017~2020년 가을 합계

```python
singer = SingerCount_Fall['합계'].sort_index()

plt.figure(figsize=(10,8))
plt.scatter(x=singer.index, y=singer, s=singer*30, c=singer, alpha=0.5)

# 상위 20위권 가수만 텍스트 출력
for i, label in enumerate(SingerCount_Fall['합계'][:20].index):
    plt.text(SingerCount_Fall['합계'][:20].index[i],SingerCount_Fall['합계'][:20][i], label)
    
plt.ylim(0,80)
plt.title('2017~2020년 가을 가수 분포')
plt.ylabel('가수별 차트인 횟수')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
```
