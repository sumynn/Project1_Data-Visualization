## 데이터 추출

```python
import pandas as pd

# 파일 읽기
month_2017 = pd.read_excel('./data/2017_month.xlsx')
month_2018 = pd.read_excel('./data/2018_month.xlsx')
month_2019 = pd.read_excel('./data/2019_month.xlsx')
month_2020 = pd.read_excel('./data/2020_month.xlsx')
month_2021 = pd.read_excel('./data/2021_month.xlsx')
```

```python
# 2017~2020 가을(9, 10, 11월) 데이터 추출
Fall = pd.DataFrame()
for data in [month_2017, month_2018, month_2019, month_2020]:
    Fall = pd.concat([Fall, data[data['월'].isin([9, 10, 11])]], ignore_index=True)
Fall
```

```python
# 연도별 가을(9~11월) 데이터 추출
Fall_2017 = Fall[Fall['년도']==2017]
Fall_2018 = Fall[Fall['년도']==2018]
Fall_2019 = Fall[Fall['년도']==2019]
Fall_2020 = Fall[Fall['년도']==2020]
```

## 제목 길이 분석

```python
import matplotlib.pyplot as plt
# 한글 표시
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False

# 색상 지정
colors = ['indianred', 'coral', 'khaki', 'lightgreen', 'lightblue', 'steelblue', 'plum', 'lightgrey', 'darkseagreen', 'mediumaquamarine', 'lightpink']
```

### 가을

```python
# 2017년
title_2017 = pd.DataFrame()
title = Fall_2017['타이틀']

# 부제목, 영어 제목 제외 길이가 12자 이상인 곡 추출
for i in range(len(title)):
    if (len(title[i].split(' (')[0])>=12) and (title[i].upper() == title[i].lower()):
            title_2017 = pd.concat([title_2017, Fall_2017[i:i+1]], ignore_index=True)
title_2017
```

```python
# 중복 제거
title_2017.drop_duplicates(['타이틀'], inplace=True, ignore_index=True)
title_2017
```

```python
# 2018
title_2018 = pd.DataFrame()
title = Fall_2018['타이틀'].reset_index(drop=True)

# 부제목, 영어 제목 제외 길이가 12자 이상인 곡 추출
for i in range(len(title)):
    if (len(title[i].split(' (')[0])>=12) and (title[i].upper() == title[i].lower()):
            title_2018 = pd.concat([title_2018, Fall_2018[i:i+1]], ignore_index=True)
           
# 중복 제거
title_2018.drop_duplicates(['타이틀'], inplace=True, ignore_index=True)
title_2018
```

```python
# 2019년
title_2019 = pd.DataFrame()
title = Fall_2019['타이틀'].reset_index(drop=True)

# 부제목, 영어 제목 제외 길이가 12자 이상인 곡 추출
for i in range(len(title)):
    if (len(title[i].split(' (')[0])>=12) and (title[i].upper() == title[i].lower()):
            title_2019 = pd.concat([title_2019, Fall_2019[i:i+1]], ignore_index=True)
           
# 중복 제거
title_2019.drop_duplicates(['타이틀'],inplace=True, ignore_index=True)
title_2019
```

```python
# 2020년
title_2020 = pd.DataFrame()
title = Fall_2020['타이틀'].reset_index(drop=True)

# 부제목, 영어 제목 제외 길이가 12자 이상인 곡 추출
for i in range(len(title)):
    if (len(title[i].split(' (')[0])>=12) and (title[i].upper() == title[i].lower()):
            title_2020 = pd.concat([title_2020, Fall_2020[i:i+1]], ignore_index=True)

# 중복 제거
title_2020.drop_duplicates(['타이틀'],inplace=True, ignore_index=True)
title_2020
```

```python
# 데이터프레임 합치기
titles = pd.concat([title_2017, title_2018, title_2019, title_2020], ignore_index=True)
titles
```

```python
# 연도별 차트인 횟수 계산 후 데이터프레임 생성
title_length = pd.DataFrame(titles.value_counts('년도'), columns=['횟수'], index=[2017,2018,2019,2020])
title_length
```

```python
# 시각화
title_length.plot(kind='bar')
plt.title('제목 길이가 12자 이상인 노래의 가을 차트인 횟수')
plt.xlabel('연도')
plt.ylabel('차트인 횟수')
```
