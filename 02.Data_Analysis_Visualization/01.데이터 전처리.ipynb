# 데이터 전처리

```python
# 장르 수정
for data in [month_2017, month_2018, month_2019, month_2020, month_2021]:
    for i in range(len(data)):
        # 'OST/드라마', 'OST/해외영화' -> OST로 통일
        if (data['장르'][i].startswith('OST')) or (data['장르'][i].startswith(' 드라마')) or (data['장르'][i].startswith(' 해외영화')):
            data['장르'][i] = 'OST'
            
        # '가요/..', 'POP/..' 등 뒤에 써진 세부 장르로 변경
        else:
            data['장르'][i] = data['장르'][i].split()[-1]
```
